# Bike Share

**Histogram**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv("bike_share.csv")

# Select columns for which you want to create histograms
selected_columns = ['temp', 'atemp', 'hum', 'windspeed']

# Set the style for seaborn (optional but enhances visual appeal)
sns.set(style="whitegrid")

# Create subplots for each selected column
plt.figure(figsize=(15, 8))
for i, column in enumerate(selected_columns, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[column], kde=True, bins=20, color='skyblue')
    plt.title(f'Histogram of {column}')

plt.tight_layout()
plt.show()
```

**Monthly Rental Trend**

```python
# Convert 'dteday' column to datetime format
df['dteday'] = pd.to_datetime(df['dteday'])

# Extract month from 'dteday' and create a new column
df['month'] = df['dteday'].dt.month

# Group by month and sum the total rentals for each month
monthly_rentals = df.groupby('month')['total_rentals'].sum()

# Plotting the monthly rental trends
plt.figure(figsize=(10, 6))
monthly_rentals.plot(marker='o', linestyle='-', color='b')
plt.title('Monthly Rental Trends')
plt.xlabel('Month')
plt.ylabel('Total Rentals')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

**Time Series Plot**

```python
# Convert 'dteday' column to datetime format
df['dteday'] = pd.to_datetime(df['dteday'])

# Plotting the time series of total rentals
plt.figure(figsize=(15, 6))
plt.plot(df['dteday'], df['total_rentals'], marker='o', linestyle='-', color='b')
plt.title('Time Series of Total Rentals')
plt.xlabel('Date')
plt.ylabel('Total Rentals')
plt.grid(True)
plt.show()
```

**Weekday vs Weekend**

```python
# Convert 'dteday' column to datetime format
df['dteday'] = pd.to_datetime(df['dteday'])

# Create a new column 'day_type' to identify weekdays and weekends
df['day_type'] = df['dteday'].dt.day_name()
df['day_type'] = df['day_type'].apply(lambda x: 'Weekend' if x in ['Saturday', 'Sunday'] else 'Weekday')

# Plotting the analysis of weekday vs. weekend rentals
plt.figure(figsize=(10, 6))
sns.boxplot(x='day_type', y='total_rentals', data=df, hue='day_type', dodge=False)
plt.title('Weekday vs. Weekend Rentals')
plt.xlabel('Day Type')
plt.ylabel('Total Rentals')
plt.show()
```

**Seasonal Patterns**

```python
# Create a box plot to show the distribution of rentals across seasons
plt.figure(figsize=(10, 6))
sns.boxplot(x='season', y='total_rentals', data=df, hue='season', dodge=False)
plt.title('Seasonal Patterns in Bike Rentals')
plt.xlabel('Season')
plt.ylabel('Total Rentals')
plt.xticks([0, 1, 2, 3], ['Spring', 'Summer', 'Fall', 'Winter'])
plt.show()
```

**Correlation Heatmap**

```python
# Select relevant numerical columns for the correlation heatmap
numerical_columns = ['temp', 'atemp', 'hum', 'windspeed', 'casual', 'registered', 'total_rentals']

# Calculate the correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```

# Daily Show Guests Cleaned

**Guests per Show Type**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv("daily_show_guests_cleaned.csv")

plt.figure(figsize=(10, 6))
df['Group'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Guests per Show Type')
plt.xlabel('Show Type')
plt.ylabel('Number of Guests')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Guest per Year**

```python
# Convert the 'Show' column to datetime format
df['Show'] = pd.to_datetime(df['Show'])

# Extract the year from the 'Show' column
df['Year'] = df['Show'].dt.year

# Create a bar chart for the number of guests per year
plt.figure(figsize=(10, 6))
df['Year'].value_counts().sort_index().plot(kind='bar', color='skyblue')
plt.title('Number of Guests per Year')
plt.xlabel('Year')
plt.ylabel('Number of Guests')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Top Guest or Occupation**

```python
top_guests = df['GoogleKnowlege_Occupation'].value_counts().nlargest(10)  # Change the number as needed
plt.figure(figsize=(12, 6))
top_guests.plot(kind='bar', color='skyblue')
plt.title('Top Occupations of Show Guests')
plt.xlabel('Occupation')
plt.ylabel('Number of Appearances')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Insurance Premiums

**Regional Comparison**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv("insurance_premiums.csv")

plt.figure(figsize=(12, 6))
sns.barplot(x='Region', y='fatal_collisions', data=df, hue='Region', dodge=False, palette='viridis')
plt.title('Fatal Collisions by Region')
plt.xlabel('Region')
plt.ylabel('Fatal Collisions')
plt.show()
```

**Premiums and Insurance Losses**

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='premiums', y='insurance_losses', data=df, hue='Region', palette='viridis')
plt.title('Premiums vs. Insurance Losses')
plt.xlabel('Premiums')
plt.ylabel('Insurance Losses')
plt.show()
```

**Fatal Collision Rates by State**

```python
plt.figure(figsize=(15, 6))
sns.barplot(x='State', y='fatal_collisions', data=df, hue='State', dodge=False, palette='viridis')
plt.title('Fatal Collision Rates by State')
plt.xlabel('State')
plt.ylabel('Fatal Collisions')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# US Counties

**Time Series of COVID-19 Cases and Deaths**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a DataFrame
df = pd.read_csv("us-counties.csv", encoding='ISO-8859-1')

# Convert the 'date' column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Create a time series plot for COVID-19 cases and deaths using seaborn
plt.figure(figsize=(12, 6))
sns.lineplot(x='date', y='cases', data=df, label='Cases', marker='o')
sns.lineplot(x='date', y='deaths', data=df, label='Deaths', marker='o')
plt.title('COVID-19 Cases and Deaths Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```
